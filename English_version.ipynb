**Comparing the median of each season**

We obtained the following results: in winter, the median number of bicycles rented is 632; in autumn, 898; in spring, 823; and in summer, the highest median is 1,214 bicycles rented. These numbers highlight the variations in demand throughout the seasons of the year.

Now, we can use a statistical test to compare spring and autumn to determine if they are statistically different from each other.

**Mann-Whitney Test**

We will apply a test called Mann-Whitney, which is a non-parametric test for independent samples, to determine if the distributions between autumn and spring are equal or different.

 Formulating the hypotheses:

$H_0:$ The distributions are equal.\
$H_1:$ The distributions are different.

```python
from scipy.stats import mannwhitneyu
```

 This Mann-Whitney test calculates two statistics at the end. One is the u value, which is the comparison between the two distributions we are trying to compare. The other is the p value. This p value indicates whether something has the statistical significance we are looking to discover.

```python
# Filtering the spring and autumn data
primavera = df_limpo[df_limpo['estacao'] =='Primavera']['contagem']
outono = df_limpo[df_limpo['estacao'] == 'Outono']['contagem']

# Applying the Mann-Whitney test
u_statistic, p_value = mannwhitneyu(primavera, outono, alternative='two-sided')
u_statistic, p_value
```

Our p_value is 0.00047, which is less than 0.05. When the p is less than 0.05, we reject the null hypothesis and accept the alternative hypothesis. In other words, the distributions are indeed different. Therefore, the idea is for the company to have different action plans for each season.

### **Analyzing demand by day and month**

We have already analyzed the data by time and obtained some interesting insights. Now, the idea is to try to understand how the bicycle count is distributed over time, considering both days and months. Therefore, the goal is to create a chart that allows us to analyze this evolution over time.

We have the date column in our DataFrame. However, remember that we had a time component that we removed to create another column. This means we will have multiple repeated dates. Therefore, the idea for this analysis will be to group the data by date, sum the bicycle count for each date, and plot this in some way.

```python
df_tempo=df_data.groupby('data')['contagem'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=df_tempo, x='data', y='contagem')
plt.title('Número de alugueis de biciccletas por datas')
plt.xlabel('Data')
plt.ylabel('Número de alugueis')
plt.show()
```

Our x-axis spans from January 2015 to January 2017. When we have data distributed over time like this, we call it a time series. With it, we understand the evolution of data over time—in this case, the number of rentals.

We can already derive some interesting insights from this chart.

If we look closely, at the beginning of 2015, the count seems to be lower and gradually increases until mid-year, around June and July, and then drops again until the end of the year. This pattern repeats for the following year, 2016.

But there are some curious details. The year 2015 has two huge peaks between July and August, significantly higher than the rest of the chart. What could have happened in London during these times to cause such a large peak?

## **Building the Predictive Model**

Now, let's predict how many bicycles will be rented the next day, the following month, and so on.

To do this, we'll use a library called Prophet. Previously known as Facebook Prophet, it was developed by Facebook and is now part of Meta.

```python
from prophet import Prophet
```

```python
# Ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df_prophet= df_data[['data','contagem']].rename(columns={'data':'ds','contagem':'y'}) # Prophet reconhece a data como ds be a variável a ser prevista como y
df_prophet.head()
```

However, an important detail is that, since we removed the times from the dates, we have repeated days in our dataframe. We need to aggregate this information to have only one record per day, with the sum of the counts for that day.

```python
df_prophet= df_prophet.groupby('ds')['y'].sum().reset_index()
df_prophet.head()
```

**Setting the Random Seed**

Now, we will use the Prophet library. To ensure that everyone gets the same results, we will set a random seed using the NumPy library. This ensures the reproducibility of the results.

```python
# Setting a Seed to Ensure Reproducibility
import numpy as np
np.random.seed(4587)

# Instantiating the Prophet Model
modelo = Prophet()

# Training the Model
modelo.fit(df_prophet)

# Creating a DataFrame for Future Predictions
futuro = modelo.make_future_dataframe(periods=90, freq='D')

# Making the Prediction
previsão = modelo.predict(futuro)
```

Additionally, there is a lighter blue band surrounding the forecast line. This band is our confidence interval. It is important that the black dots, our actual y values, fall within this confidence interval.

When you execute it, you will see three graphs in the same image. Each graph represents a different component that Prophet used to make the prediction.

The first component is the trend. It indicates whether the forecast is increasing or decreasing. Here, it shows that the forecast is trending upwards.

We also have two components related to seasonality, as discussed in the previous lesson. One of these components is weekly seasonality, and we can observe how rentals decrease on weekends. The graph shows values for each day of the week, where we see that rentals are lower on Sundays, increase on Monday and Tuesday, remain stable from Wednesday to Thursday, and decrease again from Thursday to Friday and from Friday to Saturday.

Another component considered for the forecast is annual seasonality. We can see that rentals are lower in January, gradually increase until the middle of the year, reach a peak, and then decrease again towards the end of the year.

The Prophet library uses an additive model that includes seasonal and trend components. The equation used by Prophet to make predictions can be represented as follows:

$y(t) = g(t) + s(t) + h(t) + \epsilon_t$

Onde:

- $ y(t)$  is the observed value in the time series at time \( t \).
- $ g(t) $  is the trend component that models long-term changes.
- $ s(t) $  is the seasonal component that models periodic or seasonal patterns.
- $ h(t) $ is the holiday effect component.
- $\epsilon_t$  is the random error in the prediction.

Each of these components is modeled separately. The trend component, $ g(t)$,is often modeled as a logistic or polynomial function. The seasonal component, $s(t)$, is modeled using Fourier analysis to capture periodic patterns. The holiday effect component, $h(t)$, is modeled as an additive effect for each specified holiday.

Thus, the complete equation used by Prophet for making predictions is the sum of these components plus the random error.

> Although in our project we didn't detail the inclusion of holidays in the model, it's important to highlight that this possibility exists.


### **Separating the Data into Training and Testing Sets**

Separating data into training and testing sets is crucial in time series models to evaluate the model's generalization ability. Imagine you are training a model to predict the number of sales for a company. If you train the model with all available data and then test it on the same data, it might seem excellent, but in practice, it could fail to predict future outcomes.

This occurs because the model might simply memorize the specific patterns in the training data without truly learning to identify general patterns that apply to new data. By splitting the data into training and testing sets, you simulate the scenario of future prediction, ensuring that the model is assessed on its generalization ability, or its capability to make accurate predictions on data it has never seen before. This helps to ensure that the model is robust and reliable when used for making real-world predictions.

### Obtaining metrics

**Model Performance Evaluation Metric**

Now, we need to understand how poorly the model is performing. For this, we'll use an appropriate metric.

When working with linear regression, we often use the R² metric to evaluate model performance. This metric is suitable for linear relationships.

However, in time series, we don't have a simple linear relationship, as values can fluctuate up and down due to various factors like seasonality. Therefore, the R² metric is not the best choice here.

The idea is to use a metric appropriate for time series, which compares the predicted values with the actual values and provides an estimate of how accurate the predictions are. We want to know if we are, on average, missing the count of bicycles by a certain amount up or down. baixo.

```python
df_previsao = previsao[['ds','yhat']]
df_comparacao = pd.merge(df_previsao, df_teste, on='ds') # merge dataframes 
df_comparacao
```

We noted that, on the first date of the test period, August 10, 2016, the actual value of 
𝑦 was 38,430 bicycles, while the forecast (𝑦ℎ𝑎𝑡) was 40,646 bicycles. The prediction was relatively close to the actual value.

However, for other dates, especially towards the end, the predicted values are quite far from the actual values. For example, on December 30, 2016, the actual value was 11,500 bicycles, but the forecast was 54,505 bicycles, which is a significant difference.


Mean Squared Error (MSE) is a common performance evaluation metric for regression and forecasting models. It calculates the average of the squares of the differences between the values predicted by the model and the actual values observed in the test data. Mathematically, MSE is calculated using the formula:

$MSE = \frac{\sum_{} (y_i - \hat{y}_i)^2}{n}$

However, MSE has a limitation: its units are squared relative to the original units of the data. This can make it difficult to interpret the value. To overcome this limitation, it is often preferable to use RMSE (Root Mean Squared Error). RMSE is simply the square root of MSE and is calculated using the formula:

$RMSE = \sqrt{MSE}$

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(df_comparacao['y'],df_comparacao['yhat'])
rmse = np.sqrt(mse)

print(f'MSE:{mse},RMSE:{rmse}')
```

Note that the MSE resulted in a very high value, as expected, due to the squaring of the differences. However, RMSE is more understandable, and in this case, it was 21,590. This means that our model is, on average, off by 21,478 bicycles per day, which is a quite significant error.
